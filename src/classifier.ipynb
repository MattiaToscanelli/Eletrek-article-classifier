```python
def create_vocabulary(document, remove_stop_words=False, remove_punctuation=False, remove_numbers=False, docLanguage='english'):
    
    tokens = nltk.word_tokenize(document, language=docLanguage)
    stop_words = set(nltk.corpus.stopwords.words(docLanguage)) if remove_stop_words else []
    punctuation = set(string.punctuation) if remove_punctuation else []

    vocabulary = [t.lower() for t in tokens 
                  if not ((t.lower() in stop_words)
                  or (t.lower() in punctuation)
                  or (t.lower().isdigit() and remove_numbers))]
    
    return list(set(vocabulary))
```

```python
import re


class NaiveBayes:
    @staticmethod
    def get_posterior_probabilities(training_data, verbose=True):

        posterior = {}
        # vocabulary = get_vocabulary(training_data['vocab'])
        vocabulary = training_data['vocab']
        lw = len(vocabulary)
        classes = list(set(training_data['label']))

        for index, c in enumerate(classes):
            tokens = sum(training_data['vocab'][training_data['label'] == c], [])
            try:
                den = len(list(set(tokens)))
            except:
                den = 0

            current_class_docs = tokens

            for w in vocabulary:
                num = current_class_docs.count(w)
                posterior[(w, c)] = (1 + num) / (den + lw)

                if verbose:
                    print('_' * 30)
                    message = 'Token ' + w + ' compare ' + str(num) + ' volte nella classe ' + c
                    message = re.sub('1 times', 'once', message)
                    print(message)

                    message = 'Ci sono ' + str(den) + ' tokens in classe ' + c
                    message = re.sub('are 1 tokens', 'is 1 token', message)
                    print(message)

                    print(current_class_docs)
                    print('Vocaboli: ' + str(lw))
                    print('Posterior senza Laplace smoothing: ' + str(num) + '/' + str(den) + '=' + str(round(num / den, 2)))
                    print('Posterior con Laplace smoothing: ' + str(1 + num) + '/' + str(den + lw) + '=' + str(round(posterior[(w, c)], 2)))

        return posterior
```

```python
from sklearn.model_selection import train_test_split

# split solar and ebikes datasets into train and test sets
solar_train, solar_test = train_test_split(solar, test_size=0.2, random_state=42)
ebikes_train, ebikes_test = train_test_split(ebikes, test_size=0.2, random_state=42)
```

```python
train = pd.concat([solar_train, ebikes_train])
test = pd.concat([solar_test, ebikes_test])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
# Extract the content and label columns from the dataframes
train_data = train["content"].values
train_labels = train["label"].values
test_data = test["content"].values
test_labels = test["label"].values


# Conversione dei testi in features (conteggio parole)
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(train_data)
test_features = count_vect.transform(test_data)

# Training 
classifier = MultinomialNB()
# classifier.fit(train_data, train_labels)
classifier.fit(X_train_counts, train_labels)

# Test
predictions = classifier.predict(test_features)

from sklearn.metrics import classification_report
target_names = ['solar', 'ebikes']
print(classification_report(test_labels, predictions, target_names=target_names))
```
